```python
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
work_dir = "C:/Projects/univer/6 term/coursework-ml"
```

```python
alpha = [1, 0.95, 0.9, 0.75, 0.5, 0.01]
beta = [0.01, 0.05, 0.1, 0.25, 0.5, 0.99]

image_names = [f'500_{a}_{b}.png' for a, b in zip(alpha, beta)]
images = [Image.open(os.path.join(work_dir, image_name)) for image_name in image_names]
```

```python
rows, cols = 3, 2
fig, axes = plt.subplots(rows, cols, figsize=(8, 12))

k = 0

for row in range(rows):
    for col in range(cols):
        axes[row][col].imshow(images[k])
        axes[row][col].set_title(f'alpha = {alpha[k]}, beta = {beta[k]}')
        k += 1
        
fig.tight_layout()
fig.savefig('Comparison based on alpha and beta.png')
plt.show()
```

```python
image_names = ['500_1_0.01.png', '500_normalized.png']
images = [Image.open(os.path.join(work_dir, image_name)) for image_name in image_names]
```

```python
rows, cols = 1, 2

fig, axes = plt.subplots(rows, cols)

axes[0].imshow(images[0])
axes[0].set_title('Unnormalized image')
axes[1].imshow(images[1])
axes[1].set_title('Normalized image')

fig.tight_layout()
fig.savefig('Comparison based on normalization.png')
plt.show()
```
